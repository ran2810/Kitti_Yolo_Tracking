```python
from ultralytics import YOLO
import os
```

```python
yaml_file_path = os.path.join(os.path.dirname(os.getcwd()), 'kitti.yaml')
```

```python
# Set the device to GPU or CPU
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
model = YOLO("yolov8n.pt")  # or yolov8s.pt

model.train(
        data=yaml_file_path,
        epochs=10,
        imgsz=320,
        batch=4,
        workers=0,
        augment=False
    )
```
